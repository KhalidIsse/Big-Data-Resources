#### Generate a random array with 5 elements and display it

```python
import numpy as np

# Generate 5 random numbers
vec = np.random.randint(0, 100, size=5)
print("Original vector:", vec)
```

#### Sort the array using straight insertion

```python
# Sorting the vector using insertion sort
for j in np.arange(1, vec.size):
    v = vec[j]
    i = j
    while i > 0 and vec[i-1] > v:
        vec[i] = vec[i-1]
        i = i - 1
    vec[i] = v

print("Sorted vector:", vec)
```

```python
from datetime import datetime, timedelta


np.random.seed(0)


num_rows = 100



data = {
    'State': np.random.choice(['CA', 'NY', 'TX', 'FL', 'MN'], num_rows),
    'Location': np.random.choice(['CityA', 'CityB', 'CityC'], num_rows),
    'Date Posted': [datetime.now() - timedelta(days=int(i)) for i in np.random.randint(0, 100, num_rows)],
    'Date Collected': [datetime.now() - timedelta(days=int(i)) for i in np.random.randint(0, 100, num_rows)],
    'Sample Type': np.random.choice(['Drinking Water', 'Surface Water', 'Air Filter'], num_rows),
    'Unit': np.random.choice(['pCi/L', 'Bq/L'], num_rows),
    'Cs-134': np.random.choice([np.nan, 0.05, 0.15, 0.25], num_rows),  # Random values with some NaNs
    'I-131': np.random.choice([np.nan, 0.10, 0.20, 0.30], num_rows),
    'Te-129': np.random.choice([np.nan, 0.01, 0.02, 0.03], num_rows)
}


mock_df = pd.DataFrame(data)


mock_df.head()
```

```python
# Convert the data type of specific columns
mock_df['Date Posted'] = pd.to_datetime(mock_df['Date Posted'])
mock_df['Date Collected'] = pd.to_datetime(mock_df['Date Collected'])
mock_df['State'] = mock_df['State'].astype('category')

# Checking types
mock_df.dtypes
```

### Exercise 6: Aggregation and Grouping Data

```python
# Grouping by 'State' and calculating the mean of Cs-134
grouped = mock_df.groupby('State')['Cs-134'].mean()
print(grouped)
```

```python
# Applying multiple aggregation functions
agg_data = mock_df.groupby(['State', 'Location']).agg({'Cs-134': ['mean', 'std'], 'Te-129': ['min', 'max']})
print(agg_data)
```

### Exercise 7: Exporting Data in Different Formats

### Exercise 9: Creating a Graph

```python
x = np.linspace(0, 100, 500)
y = np.sin(2 * np.pi * x / 100)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.title("Sine Wave")
plt.show()
```
